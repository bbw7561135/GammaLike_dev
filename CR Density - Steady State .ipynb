```python
def plotCR(fits1, fits2, ebin, species, sp, nrow, ncol=4, projection='plane', text='', text2=''):
    '''
    species: protons, leptons, positrons
    projection: 'plane', or 'column'
    '''
    #plt.subplot(nrow, ncol, sp)
    hdu = pyfits.open(fits1)
    hdu2 = pyfits.open(fits2)
    hdr = hdu[0].header
    #print hdu[0].header.cards
    
    #print hdu[0].data.shape
    
    if projection=='plane':
        d = hdu[0].data[:,:,25,:,:] # Just look at the plane 
        d2 = hdu2[0].data[:,:,25,:,:] # Just look at the plane 
    else:
        d = np.sum(hdu[0].data, axis=2) # Sum over the vertical?     
        d2 = np.sum(hdu2[0].data, axis=2) # Sum over the vertical?     
    
        
    if species=='protons':
        d = d[3, ebin, :,:]
        d2 = d2[3, ebin, :,:]
    if species=='leptons':
        d = np.sum(d[0:3, ebin, :,:], axis=0)
        d2 = np.sum(d2[0:3, ebin, :,:], axis=0)
    if species=='positrons':
        d = d[0, ebin, :,:]
        d2 = d2[0, ebin, :,:]
    
    im = plt.imshow(np.log10(d2/d), extent=[-20,20,-20,20], origin='lower')
    cbar = plt.colorbar(im)
    
    E = 10**hdr['CRVAL4'] * 10**(hdr['CDELT4']*ebin) # Energy in MeV
    plt.text(.05, 0.05, '%1.1f GeV'%(E/1e3), transform=plt.gca().transAxes, color='w')
    plt.text(.95, 0.05, text, transform=plt.gca().transAxes, color='w', horizontalalignment='right')
    plt.text(.05, 0.9, text2, transform=plt.gca().transAxes, color='w', horizontalalignment='left')
    plt.xlim(-15,15)
    plt.ylim(-15,15)
    plt.scatter(8.5,0, marker='+',s=100, color='w')
    
    
    print d.shape
    
    return im 
    
plt.figure(figsize=(8,6))
plotCR('/data/GCE_sys/new/nuclei_full_54_mod_s_42', '/data/GCE_sys/new/nuclei_full_54_mod_s_46',ebin=29,species='protons', sp=1,nrow=2,ncol=2, projection='plane', text='$p$')
plotCR('/data/GCE_sys/new/nuclei_full_54_mod_s_42', '/data/GCE_sys/new/nuclei_full_54_mod_s_46',ebin=29,species='leptons', sp=2,nrow=2,ncol=2, projection='plane', text=r'$e^+ + e^-$')

# plotCR('/data/GCE_sys/new/nuclei_full_54_mod_s_46',ebin=51,species='protons', sp=3,nrow=2,ncol=2, projection='plane', text='$p$')
# plotCR('/data/GCE_sys/new/nuclei_full_54_mod_s_46',ebin=51,species='leptons', sp=4,nrow=2,ncol=2, projection='plane', text=r'$e^+ + e^-$')
```
