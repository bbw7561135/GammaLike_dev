```python
import h5py

best_fit = -10

plt.figure(figsize=(8,3))
plt.subplot(121)
for i in range(0,10,1):
    
    h5 = h5py.File('../mod_c/mod_f_'+str(i)+'_stripped.hdf5', 'r')
#     for f in h5['/fit_results/']:
#         print f
        
        
        
    try:    
        ll =  h5['/fit_results/scan_axesratio/'].attrs['loglike_total']
        ar =  h5['/fit_results/scan_axesratio/'].attrs['axesratio']
        #print ar, ll

        if i == best_fit:
            plt.plot(ar,2*(ll-min(ll)), c='k', alpha=1)
        else:
            plt.plot(ar,2*(ll-min(ll)), c='goldenrod', alpha=(i)/10., label=str(i))
      
    except:
        pass
    
plt.legend(frameon=False)
plt.ylabel(r'$2\Delta\log{(\mathcal{L})}$')
plt.xlabel(r'Axes Ratio ($R<1$ = disklike)')
plt.text(.1,.9, 'New Gas Models', transform=plt.gca().transAxes)
plt.ylim(0,50)
plt.xlim(.6,1.5)


plt.subplot(122)
for i in range(10,20,2):
    
    h5 = h5py.File('../mod_c/mod_f_'+str(i)+'_stripped.hdf5', 'r')
#     for f in h5['/fit_results/']:
#         print f
        
        
        
    try:    
        ll =  h5['/fit_results/scan_axesratio/'].attrs['loglike_total']
        ar =  h5['/fit_results/scan_axesratio/'].attrs['axesratio']
        #print ar, ll

        if i == best_fit:
            plt.plot(ar,2*(ll-min(ll)), c='k', alpha=1)
        else:
            plt.plot(ar,2*(ll-min(ll)), c='goldenrod', alpha=(i-10+1)/10., label=str(i))
      
    except:
        pass
    
plt.legend(frameon=False)
plt.ylabel(r'$2\Delta\log{(\mathcal{L})}$')
plt.xlabel(r'Axes Ratio ($R<1$ = disklike)')
plt.ylim(0,2e1)
plt.text(.1,.9, 'Old Gas Models', transform=plt.gca().transAxes)
plt.xlim(.6,1.5)
plt.savefig('/home/carlson/Dropbox/plots/GCE_sys/scan_axis_ratio.pdf', bbox_inches='tight')
```
