```python
import h5py

best_fit = 12

min_gam = []

for i in range(0,65):
    
    h5 = h5py.File('../mod_c/mod_d_'+str(i)+'_stripped.hdf5', 'r')
#     for f in h5['/fit_results/scan_gamma/'].attrs:
#         print f
    try:    
        ll =  h5['/fit_results/scan_gamma/'].attrs['loglike_total']
        gam =  h5['/fit_results/scan_gamma/'].attrs['gamma']
        
        if i == best_fit:
            plt.plot(gam,2*(ll-min(ll)), c='k', alpha=1, label='Best Fit')
        else:
            plt.plot(gam,2*(ll-min(ll)), c='y', alpha=.25)
        min_gam.append(gam[np.argmin(ll)])    
    except:
        pass
print np.average(min_gam)
plt.vlines(np.average(min_gam), 0,100, 'm', label='Average Best Fit Slope')
plt.xlabel(r'Inner Slope $\alpha$')
plt.ylabel(r'$2\Delta\log{(\mathcal{L})}$')
plt.ylim(0,1e2)
plt.legend(frameon=False, fontsize=9)
plt.xlim(1,1.5)
```
