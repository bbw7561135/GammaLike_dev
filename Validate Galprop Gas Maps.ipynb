```python
#d1 = pyfits.open('/data/galprop2/output/bremss_healpix_54_NSPEB_no_secondary.gz')[1]

def ReadFits(fname):
    hdu = pyfits.open(fname)
    return np.array([hdu[1].data.field(i) for i in range(len(hdu[1].data[0]))])

#d1 = pyfits.open('/data/galprop2/output/bremss_healpix_54_base_no_secondary.gz')[1].data

b=10

import healpy
d1 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_base_single_HI.gz')
d2 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_NSPEB_single_HI.gz')
d3 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_F07_single_HI.gz')

maps = [healpy.cartview(d[b], latra=(-90,90), lonra=(-180,180), hold=True,  return_projected_map=True,xsize=1280) for i, d in enumerate([d1,d2,d3])]
plt.clf()


d1_H2 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_base_single_H2.gz')
d2_H2 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_NSPEB_single_H2.gz')
d3_H2 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_F07_single_H2.gz')

maps_H2 = [healpy.cartview(d[b], latra=(-90,90), lonra=(-180,180), hold=True,  return_projected_map=True,xsize=1280) for i, d in enumerate([d1_H2,d2_H2,d3_H2])]
plt.clf()


```

```python
plt.figure(figsize=(8,4))




for i, d in enumerate([d1,d2,d3]):
    plt.figure(figsize=(8,4))
    #plt.subplot(2,3,i+1)
    plt.imshow(np.log10(maps[i]),cmap='gist_heat', origin='lower', extent=[180,-180,-90,90])
    plt.colorbar()
    plt.show()

for i, d in enumerate([d1_H2,d2_H2,d3_H2]):
    plt.figure(figsize=(8,4))
    #plt.subplot(2,3,i+1)
    plt.imshow(np.log10(maps_H2[i]),cmap='gist_heat', origin='lower', extent=[180,-180,-90,90])
    plt.colorbar()
    plt.show()
    

# m = healpy.cartview(np.log10(d2[b]), latra=(-90,90), lonra=(-180,180), return_projected_map=True,xsize=1280)
# plt.clf()
# plt.imshow(m,cmap='gist_heat', origin='lower', vmin=0,vmax=2)
# plt.colorbar()
# plt.show()

# d1 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_base_single_HI.gz')
# d2 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_NSPEB_single_HI.gz')
# d3 = ReadFits('/data/galprop2/output/pi0_decay_healpix_54_F07_single_HI.gz')
```
